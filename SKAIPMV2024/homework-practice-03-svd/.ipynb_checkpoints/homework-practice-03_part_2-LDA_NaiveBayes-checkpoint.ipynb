````python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random samples
M = 200
sigma1 = 0.1
sigma2 = 0.5

# Generate two independent Gaussian distributions
x1 = np.random.randn(M, 1) * sigma1
x2 = np.random.randn(M, 1) * sigma2

# Concatenate them into a single dataset
X = np.concatenate((x1, x2), axis=1)

# Step 2: Rotate the data points by a given angle alpha
alpha = np.radians(45)  # For example, 45 degrees
rotation_matrix = np.array([[np.cos(alpha), -np.sin(alpha)], 
                            [np.sin(alpha), np.cos(alpha)]])

# Rotate the points
X_rotated = X @ rotation_matrix.T

# Step 3: Compute the covariance matrix of the rotated points
cov_matrix = np.cov(X_rotated, rowvar=False)

# Step 4: Generate points from a multivariate normal distribution
mean = [0, 0]  # Mean of the distribution
cov = [[sigma1**2, 0], [0, sigma2**2]]  # Covariance matrix
X_multivariate = np.random.multivariate_normal(mean, cov, M)

# Step 5: Visualize the data
plt.figure(figsize=(10, 5))

# Original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='blue', alpha=0.5, label='Original Data')
plt.title('Original Gaussian Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()

# Rotated data
plt.subplot(1, 2, 2)
plt.scatter(X_rotated[:, 0], X_rotated[:, 1], color='green', alpha=0.5, label='Rotated Data')
plt.title('Rotated Data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()

plt.tight_layout()
plt.show()

# Compare with the multivariate normal distribution
plt.figure()
plt.scatter(X_multivariate[:, 0], X_multivariate[:, 1], color='red', alpha=0.5, label='Multivariate Normal Data')
plt.title('Multivariate Normal Distribution')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()

plt.show()
````

### 2. Плотность вероятности гауссового распределения
В общем виде $n$-мерное гауссово распределение имеет вид:  
$$p(x) = \frac{1}{(\sqrt{2\pi})^n \sqrt{detC}} e^{-\frac{1}{2}(x-\mu)^TС^{-1}(x-\mu)} \tag{1}$$  
где $x\in{R^n}$, $\mu\in{R^n}$ - среднее значение, $C$ - матрица ковариации.  

Сгенерируйте набор точек в двумерном пространстве (как делали в пункте 1 или другим способом). Посчитайте для них среднее значение и матрицу ковариации $\mu$ и $C$. Для оценки плотности вероятности (probability density function - `pdf`) гауссового распределения в произвольной точке используйте `scipy.stats.multivariate_normal` 
(https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.multivariate_normal.html), либо можете посчитать вручную.  

Визуализируйте набор точек и плотность вероятности, для чего изобразите точки одним цветом, а фон раскрасьте в соответсвии с плотностью вероятности, для чего можете использовать пример ниже:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Create a grid of points in 2D space
xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
XX, YY = np.meshgrid(xx, yy)

# Stack the grid points into pairs of (x, y)
pp = np.stack((XX.flatten(), YY.flatten()), axis=1)

# Define the mean and covariance matrix for the Gaussian distribution
mean = [0, 0]
cov = [[0.3, 0], [0, 0.1]]

# Calculate the probability density function (PDF) over the grid points
m = multivariate_normal(mean=mean, cov=cov)
ZZ = m.pdf(pp).reshape(XX.shape)

# Visualize the density using a color plot and a contour plot
plt.figure(figsize=(8, 6))

# Plot the PDF as a color plot (heatmap)
IMG = plt.pcolor(XX, YY, ZZ, shading='auto')
plt.colorbar(IMG)

# Plot the contour lines
CS = plt.contour(XX, YY, ZZ, levels=[0.2, 0.4, 0.6, 0.8, 0.9], colors='k')
plt.clabel(CS, inline=1, fontsize=10)

# Add labels and title
plt.title('Gaussian PDF Visualization')
plt.xlabel('X1')
plt.ylabel('X2')

plt.show()
````

### 3. Бинарная классификация

**Задание** 
Создайте случайный датасет, состоящий из двух классов $y\in{\{0, 1\}}$, а признаки имеют две координаты. Посчитайте для каждого класса среднее значение и матрицу ковариации $\mu_0, \mu_1$ и $C_0, C_1$.

Решение задачи классификации можно свести к оценке максимальной апостериорной вероятности (Maximum a posterior - MAP):
$$y = \underset{y\in{0,1}}{argmax} p(y|x) \tag{2}$$  
которая в свою очередь через теорему Байеса равна:  
$$p(y|x) = \frac{p(x|y)p(y)}{p(x)} \tag{3}$$
В последней формуле вероятность p(y) называется априорной и не зависит от $x$ - оценить её можно просто как долю точек датасета принадлежащего заданному классу $y$. Вероятность p(x|y) называют правдоподобием (likelyhood). Обычно правдоподобие моделируется некоторой генерирующей моделью, например по формуле гаусса (1), и говорит насколько правдоподобно что точка с координатами $x$ могла быть "сгенерирована" в классе $y$. Знаменатель $p(x)$ в последней формуле не зависит от $y$, поэтому на него внимание не обращаем.  

Таким образом, для бинарной классификации, т.е. для оценки принадлежности некоторой точки $x$ к классу 0 или 1, нужно сравнить две величины:  
$$p(x|y=0)p(y=0) \overset{?}{<>} p(x|y=1)p(y=1) \tag{4}$$

**Задание**
Визуализируейте на одном рисунке датасет с точками обоих классов разными цветами. Раскрасьте фон изображения в соответствии с разностью левой и правой частей уравнения (4). Постройте на этом же рисунке разделяющую кривую, т.е. кривую соответствующую нулевой разнице между левой и правой частями. Для её построения используйте функции `plt.contour` с указанием конкретного уровня.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Generate random dataset for two classes with different means and covariances
np.random.seed(42)  # For reproducibility

# Class 0: mean and covariance
mean_0 = [0, 0]
cov_0 = [[1, 0.5], [0.5, 1]]
data_0 = np.random.multivariate_normal(mean_0, cov_0, 100)

# Class 1: mean and covariance
mean_1 = [2, 2]
cov_1 = [[1, -0.3], [-0.3, 1]]
data_1 = np.random.multivariate_normal(mean_1, cov_1, 100)

# Combine data from both classes
X = np.vstack((data_0, data_1))
y = np.hstack((np.zeros(100), np.ones(100)))  # Labels: 0 for class 0, 1 for class 1

# Define the Gaussian distributions for each class
dist_0 = multivariate_normal(mean=mean_0, cov=cov_0)
dist_1 = multivariate_normal(mean=mean_1, cov=cov_1)

# Create a grid of points to evaluate posterior probabilities
xx, yy = np.meshgrid(np.linspace(-3, 5, 100), np.linspace(-3, 5, 100))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Calculate likelihoods (p(x|y)) for both classes
likelihood_0 = dist_0.pdf(grid_points).reshape(xx.shape)
likelihood_1 = dist_1.pdf(grid_points).reshape(xx.shape)

# Define prior probabilities (assume equal priors for simplicity)
prior_0 = 0.5
prior_1 = 0.5

# Calculate posteriors (p(y=0|x) and p(y=1|x)) using Bayes' Theorem
posterior_0 = likelihood_0 * prior_0
posterior_1 = likelihood_1 * prior_1

# Visualize the dataset, posterior probabilities, and decision boundary
plt.figure(figsize=(8, 6))

# Plot the dataset points for both classes
plt.scatter(data_0[:, 0], data_0[:, 1], color='blue', label='Class 0')
plt.scatter(data_1[:, 0], data_1[:, 1], color='red', label='Class 1')

# Plot the background colors according to the posterior
plt.contourf(xx, yy, posterior_0 - posterior_1, levels=0, cmap="coolwarm", alpha=0.5)

# Plot the decision boundary (where posterior probabilities are equal)
plt.contour(xx, yy, posterior_0 - posterior_1, levels=[0], colors='black')

# Add labels, title, and legend
plt.title("Binary Classification: Posterior and Decision Boundary")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.colorbar()
plt.show()
````

### 4. LDA (linear discriminant analysis)

**Задание** Рассмотрите частный случай, когда для точек двух классов матрицы ковариации совпадают, т.е. $C_0=C_1$, но средние значения отличаются $\mu_0\neq\mu_1$. Выпишите явное выражение для разделяющей поверхности в этом случае. Визуализируйте такой случай аналогично предыдущему пункту. Про линейный и квадратичный дискриминантный анализ можете почитать здесь (https://scikit-learn.org/stable/modules/lda_qda.html) или в учебнике Яндекса по ML (https://education.yandex.ru/handbook/ml/article/generativnyj-podhod-k-klassifikacii)

Реализуйте классификатор на основе метода линейного дискриминантного анализа по шаблону ниже. Данный алгоритм предполагает одинаковость матриц ковариации (т.е. считает одну общую матрицу ковариации, но разные средние) и использует линейное правило классификации. Желательно сделать реализацию, которая подходит не только для двумерного, но и для более общего случая.

```python
from sklearn.base import BaseEstimator

class myLDA(BaseEstimator):
    def __init__(self):
        pass
    
    def fit(self, X, y):
        pass
    
    def predict(self, X):
        pass
```


````python
import numpy as np
from sklearn.base import BaseEstimator

class myLDA(BaseEstimator):
    def __init__(self):
        self.means = None
        self.covariance = None
        self.class_priors = None
        self.classes = None

    def fit(self, X, y):
        # Get unique classes
        self.classes = np.unique(y)
        n_classes = len(self.classes)
        n_features = X.shape[1]
        
        # Initialize means, covariance and priors
        self.means = np.zeros((n_classes, n_features))
        self.covariance = np.zeros((n_features, n_features))
        self.class_priors = np.zeros(n_classes)
        
        # Compute mean, covariance matrix and priors
        for idx, c in enumerate(self.classes):
            X_c = X[y == c]
            self.means[idx, :] = X_c.mean(axis=0)
            self.class_priors[idx] = X_c.shape[0] / X.shape[0]
        
        # Shared covariance matrix (for all classes)
        self.covariance = np.cov(X, rowvar=False)

    def predict(self, X):
        y_pred = [self._predict_single(x) for x in X]
        return np.array(y_pred)

    def _predict_single(self, x):
        # Compute discriminant scores for each class
        discriminants = []
        
        for idx, c in enumerate(self.classes):
            mean_diff = x - self.means[idx]
            inverse_cov = np.linalg.inv(self.covariance)
            discriminant = -0.5 * mean_diff.T @ inverse_cov @ mean_diff
            discriminant += np.log(self.class_priors[idx])
            discriminants.append(discriminant)
        
        return self.classes[np.argmax(discriminants)]

# Example of using the myLDA classifier
if __name__ == "__main__":
    from sklearn.datasets import make_classification
    from sklearn.model_selection import train_test_split

    # Create a sample dataset
    X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Initialize and train myLDA classifier
    clf = myLDA()
    clf.fit(X_train, y_train)

    # Make predictions
    y_pred = clf.predict(X_test)

    # Evaluate performance
    from sklearn.metrics import accuracy_score
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
````

### 5. Наивный байесовый классификатор (NaiveBayes)

**Задание** 
Прочитать по наивный байесовый подход можно здесь (https://scikit-learn.org/stable/modules/naive_bayes.html) или в учебнике Яндекса по ML (https://education.yandex.ru/handbook/ml/article/generativnyj-podhod-k-klassifikacii).  

Суть наивного предположения состоит в том, чтобы не считать совместную плотность вероятности для $n$-мерного случая, а рассматривать каждую переменную как независимую и считать одномерные распределения:  
$$p(x_1,x_2,...,x_n|y) => \prod\limits_{i=1}^{n}p(x_i|y)$$  
Это можно также рассматривать как обнуление всех внедиагональных элементов матрицы ковариации.  

**Задание** 
Реализуйте классификатор на основе наивного байесового подхода с одномерным гауссовым распределением для каждой переменной. Желательно сделать реализацию, которая подходит не только для двумерного, но и для более общего случая.

```python
from sklearn.base import BaseEstimator

class myNB(BaseEstimator):
    def __init__(self):
        pass
    
    def fit(self, X, y):
        pass
    
    def predict(self, X):
        pass
```

````python
import numpy as np
from sklearn.base import BaseEstimator

class myNB(BaseEstimator):
    def __init__(self):
        self.class_priors = None
        self.means = None
        self.variances = None
        self.classes = None

    def fit(self, X, y):
        # Get the unique classes
        self.classes = np.unique(y)
        n_classes = len(self.classes)
        n_features = X.shape[1]
        
        # Initialize means, variances and priors
        self.means = np.zeros((n_classes, n_features))
        self.variances = np.zeros((n_classes, n_features))
        self.class_priors = np.zeros(n_classes)
        
        # Calculate mean, variance, and priors for each class
        for idx, c in enumerate(self.classes):
            X_c = X[y == c]
            self.means[idx, :] = X_c.mean(axis=0)
            self.variances[idx, :] = X_c.var(axis=0) + 1e-9  # small value to avoid division by zero
            self.class_priors[idx] = X_c.shape[0] / X.shape[0]

    def predict(self, X):
        y_pred = [self._predict_single(x) for x in X]
        return np.array(y_pred)

    def _predict_single(self, x):
        # Calculate the posterior for each class
        posteriors = []
        
        for idx, c in enumerate(self.classes):
            prior = np.log(self.class_priors[idx])
            class_conditional = np.sum(np.log(self._gaussian_probability(idx, x)))
            posterior = prior + class_conditional
            posteriors.append(posterior)
        
        return self.classes[np.argmax(posteriors)]

    def _gaussian_probability(self, class_idx, x):
        mean = self.means[class_idx]
        var = self.variances[class_idx]
        numerator = np.exp(-(x - mean) ** 2 / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator

# Example of using the myNB classifier
if __name__ == "__main__":
    from sklearn.datasets import make_classification
    from sklearn.model_selection import train_test_split

    # Create a sample dataset
    X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Initialize and train myNB classifier
    clf = myNB()
    clf.fit(X_train, y_train)

    # Make predictions
    y_pred = clf.predict(X_test)

    # Evaluate performance
    from sklearn.metrics import accuracy_score
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
````
